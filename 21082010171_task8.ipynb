```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0',
    'Accept-Encoding': 'gzip, deflate',
    'Accept': 'text/html, application/xhtml+xml, application/cml;q=0.9,*/*;q=0.8',
    'DNT': '1',
    'Connection': 'close',
    'Upgrade-Insecure-Requests': '1'
}
```

```python
base_url='https://indeks.kompas.com/?site=all&date=2024-11-04'
```

```python
req = requests.get(base_url, headers=headers)
soup = BeautifulSoup(req.content, 'html.parser')
```

```python
get_html = soup.find_all('div', class_='articleItem-box')
```

```python
news_data = []
for html in get_html:
  news_dic = {}
  news_title = html.find('h2', class_='articleTitle')
  news_dic['News Title'] = news_title.text.strip() if news_title else 'unknown news title'

  pub_date = html.find('div', class_='articlePost-date')
  news_dic['Published'] = pub_date.text.strip() if pub_date else 'unknown published date'

  cat = html.find('div', class_='articlePost-subtitle')
  news_dic['Category'] = cat.text.strip() if cat else 'unknown category'

  news_data.append(news_dic)
```

```python
data = pd.DataFrame(news_data)
data
```

```python
data.to_json('news-kompas2.json', orient='index')
```
