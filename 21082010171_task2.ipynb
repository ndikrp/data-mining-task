```python
def sex_survival_bar(variable):
  died=data[data['Survived']==0][variable].value_counts()/len(data[data['Survived']==0])
  survived=data[data['Survived']==1][variable].value_counts()/len(data[data['Survived']==1])
  dataset=pd.DataFrame([died, survived])
  dataset.index=['Deceased', 'Alive']
  dataset.plot(kind='bar', stacked=False, title='Survival Percentage by Sex')
  dataset.plot(kind='pie', subplots=True, autopct='%1.2f%%', legend=False, stacked=False, title='Persentase')
  return dataset.head()

sex_survival_bar('Sex')
```

```python
def sex_pclass_bar(variable):
  first=data[data['Pclass']==1][variable].value_counts()/len(data[data['Pclass']==1])
  second=data[data['Pclass']==2][variable].value_counts()/len(data[data['Pclass']==2])
  third=data[data['Pclass']==3][variable].value_counts()/len(data[data['Pclass']==3])
  dataset=pd.DataFrame([first, second, third])
  dataset.index=['First Class', 'Second Class', 'Third Class']
  dataset.plot(kind='bar', stacked=False, title='Survival percentage based on Passenger Class')
  dataset.T.plot(kind='pie', subplots=True, autopct='%1.2f%%', legend=False, stacked=False, title='Survival percentage based on Passenger Class')
  return dataset.head()

sex_pclass_bar('Survived')
```

```python
def embarked_survival_bar():
    data['Embarked'] = data['Embarked'].map({'S': 'Southampton', 'C': 'Cherbourg', 'Q': 'Queenstown'})
    died = data[data['Survived'] == 0]['Embarked'].value_counts() / len(data[data['Survived'] == 0])
    survived = data[data['Survived'] == 1]['Embarked'].value_counts() / len(data[data['Survived'] == 1])
    dataset = pd.DataFrame([died, survived])
    dataset.index = ['Deceased', 'Survived']
    dataset.plot(kind='bar', stacked=False, title='Survival Percentage by Embarkation Port')
    dataset.T.plot(kind='pie', subplots=True, autopct='%1.2f%%', legend=False, stacked=False, title='Survival percentage based on Passenger Class')
    dataset.T.plot(kind='area', stacked=False, title='Survival percentage based on Passenger Class')
    return dataset.head()

embarked_survival_bar()

```

```python
def age_group_survival_bar():
    age_bins = [0, 18, 40, 60, 100]
    age_labels = ['Children', 'Young Adult', 'Adult', 'Elderly']
    data['AgeGroup'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels)
    died = data[data['Survived'] == 0]['AgeGroup'].value_counts() / len(data[data['Survived'] == 0])
    survived = data[data['Survived'] == 1]['AgeGroup'].value_counts() / len(data[data['Survived'] == 1])
    dataset = pd.DataFrame([died, survived])
    dataset.index = ['Deceased', 'Survived']
    dataset.plot(kind='bar', stacked=False, title='Survival Percentage by Age Group')
    dataset.T.plot(kind='pie', subplots=True, autopct='%1.2f%%', legend=False, stacked=False, title='Survival percentage based on Passenger Class')
    return dataset.head()

age_group_survival_bar()
```

```python
def family_survival_bar():
    data['FamilySize'] = data['SibSp'] + data['Parch']
    died = data[data['Survived'] == 0]['FamilySize'].value_counts() / len(data[data['Survived'] == 0])
    survived = data[data['Survived'] == 1]['FamilySize'].value_counts() / len(data[data['Survived'] == 1])

    dataset = pd.DataFrame([died, survived])
    dataset.index = ['Meninggal', 'Selamat']

    dataset.plot(kind='bar', stacked=False, title='Survival Percentages by Family Size')
    return dataset.head()

family_survival_bar()
```
